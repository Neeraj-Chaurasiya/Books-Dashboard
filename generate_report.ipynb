```python
# Import Libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Load CSV file
df = pd.read_csv(r"data\books_cleaned.csv")

# Preview first rows
df.head()
```

```python
# Total number of books
total_books = df.shape[0]

# Average price & rating
avg_price = df["Price"].mean()
avg_rating = df["Rating"].mean()

print("Total books :", total_books)
print(" Average price :", round(avg_price,2))
print(" Average rating :", round(avg_rating,2))
```

```python
# Price Distribution
sns.histplot(df["Price"], bins=20, kde=True)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# Top 10 Expensive Books

top10 = df.nlargest(10, "Price")
sns.barplot(x="Price", y="Title", data=top10, hue="Title", palette="Set2")
plt.title("Top 10 Most Expensive Books")
plt.xlabel("Price")
plt.ylabel("Book Title")
plt.show()
```

```python
# Average Price by Rating
avg_by_rating = df.groupby("Rating")["Price"].mean().reset_index()

sns.barplot(x="Rating", y="Price", data=avg_by_rating, hue= "Price", palette="Set1")
plt.title("Average Price by Rating")
plt.xlabel("Rating")
plt.ylabel("Average Price")
plt.show()
```

```python
# Automated Excel Repor
with pd.ExcelWriter("books_report.xlsx") as writer:
    df.to_excel(writer, sheet_name="All Books", index=False)
    top10.to_excel(writer, sheet_name="Top10 Expensive", index=False)
    avg_by_rating.to_excel(writer, sheet_name="Avg Price by Rating", index=False)

print("✅ Report generated: books_report.xlsx")
```
